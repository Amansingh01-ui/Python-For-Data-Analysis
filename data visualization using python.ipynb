### MatplotLib

MatplotLib is a plotting library for the pythone programming languages and its numerical mathematics extension Numpy.It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter,wxPython,Qt,Gtk+.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Let's take examples

x=np.arange(1,11)
y=np.arange(11,21)
```

```python
# ploting using matplotlib

plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph")
```

```python
y=x*x
```

```python
plt.plot(x,y,'y*',linestyle='dashed',linewidth=2,markersize=10)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Diagram')
```

```python
# Creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11)
y = 3*x+5
plt.title("MatplotLib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve

x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

# Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
## Subplot()
# Compute the x and y coordinates for points on sine and cosine curve

x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.

plt.subplot(2,1,1)

# Make the first plot

plt.plot(x,y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active and make the second plot

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

# Show the figure
plt.show()
```

```python
## Bar Plot


x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,color='r')
plt.bar(x2,y2,color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel(' X axis')

plt.show()
```

```python
## Histograms

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
## Box Plot using MatplotLib
# box plot help to find the outliers
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
data2=[1,2,3,4,6,8,10,15,23,28,18,90,225]
# rectangular box plot
plt.boxplot(data2,vert=True,patch_artist=True);
```

```python
## Pie Chart

labels = 'Python','C','C++','Java'
sizes = [215,130,245,210]
colors = ['yellow','lightblue','pink','green']
explode = (0.4,0,0,0) #explode first slice

# plot

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct = '%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```

```python
## MATLAB-Style Interface

x = np.linspace(0, 10, 100)

plt.figure()

plt.subplot(2,1,1) #(rows,columns,pannel number)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x));
```

## Seaborn

Distribution Plots:
1. distplot
2. joinplot
3. pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

#### Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix between two discrete dimension or even types. It is very important in Feature Selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

#### Join Plot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histrogram or a 2D density plot.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex');
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg');
```

#### Pair Plot

A "pairs plot" is also known as a scatterplot, in which one variable in the same data row is matched with another variables value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

#### Dist Plot

Dist Plot help us to check the distribution of the columns features

```python
sns.distplot(df['tip']);
```

#### Categorical Plots

Seaborn also help us in doing the analysis on categorical data points. In this section we will discus about
- boxplot
- violinplot
- countplot
- bar plot

```python
# count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
# Bar Plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
# Box plot
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
## categorize my data based on some other categories

sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
```

```python
# Violin Plot
sns.violinplot(x="total_bill",y="day",data=df,palette='rainbow')
```
